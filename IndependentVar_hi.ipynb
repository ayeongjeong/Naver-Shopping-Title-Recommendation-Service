# Data load

```python
# mongoDB to DataFrame

import pandas as pd
import numpy as np
from pymongo import MongoClient

client = MongoClient('mongodb://192.168.0.154:27017/')  # mongo 연결
mydb = client.mydb

modelDB = mydb.processed # get Collection

model_df = pd.DataFrame(list(modelDB.find()))
```

```python
model_df
```

# title의 단어 개수 세기

```python
title_lst = model_df['title'].tolist()
title_lst
```

```python
# 타이틀 단어를 하나의 리스트에 넣기

title_one = []

for i in range(len(title_lst)):
    title_one.extend(title_lst[i])

title_df = pd.DataFrame(title_one, columns = ['title'])
title_df
```

```python
# 단어 별 빈도수 계산하여 DataFrame 만들기

title_frq = title_df['title'].value_counts().reset_index(drop = False)
title_frq.columns = ['word','freqency']
title_frq
```

```python
# 정수인코딩 : 빈도 수 순서에 따라 인코딩 정수 부여

title_frq = title_frq.reset_index(drop = False)
title_frq.columns = ['encode', 'word','frequency']
title_frq
```
